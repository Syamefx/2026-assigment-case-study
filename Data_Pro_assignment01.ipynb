# **Question 5.**

# Checked and handled missing values in the dataset

```python
print("Missing values before :\n", df.isna().sum())
for col in ['age', 'income']:
    if df[col].isna().sum() > 0:
        med_val = df[col].median()
        df[col].fillna(med_val, inplace=True)
# Fixing missing values
most_common_gender = df['gender'].mode()[0]
df['gender'].fillna(most_common_gender, inplace=True)

print("\nMissing values after applying fixes:\n", df.isna().sum())
```

# **Question 6.**

#Checking the outliers and handle outliers in performance score using Percentiles


```python
# finding out the upper and lower percentile limits
lower_lim = df['performance_score'].quantile(0.05)
upper_lim = df['performance_score'].quantile(0.95)

print(f"Capping scores below {lower_lim} and above {upper_lim}")

# capping the outliers directly
df['performance_score'] = np.clip(df['performance_score'], lower_lim, upper_lim)
print(df['performance_score'])
print("-"*200)
print(df)
```
